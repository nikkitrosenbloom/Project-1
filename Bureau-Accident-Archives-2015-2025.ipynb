```python
import pandas as pd

url = "https://www.baaa-acro.com/crash-archives?created=2015-01-02&created_1=2025-06-28&field_crash_region_target_id=All&field_crash_country_target_id=&field_crash_registration_target_id=&field_crash_aircraft_target_id=&field_crash_operator_target_id=&field_crash_cause_target_id=All&field_crash_zone_target_id=&field_crash_site_type_target_id=All&field_crash_phase_type_target_id=All&field_crash_flight_type_target_id=All&field_crash_survivors_value=All&field_crash_city_target_id=&page="
df = pd.read_html(url)[0]
```

```python
df
```

```python
all_tables = []

for page_num in range(67):
    full_url = url + str(page_num)
    try:
        df = pd.read_html(full_url)[0]
        all_tables.append(df)
        print(f"Page {page_num + 1} scraped successfully.")
    except Exception as e:
        print(f"Failed to scrape page {page_num + 1}: {e}")
```

```python
final_df = pd.concat(all_tables, ignore_index=True)
```

```python
final_df
```

```python
extra_final_df = final_df.drop(columns=['Image', 'Registration', 'Unnamed: 7'])
```
