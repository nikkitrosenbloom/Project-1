# Data Analysis: Bureau of Accident Archives

### I am analyzing data from Bureau of Accident Archives, from 01/01/2015 to 06/25/2025.

```python
import pandas as pd

url = "https://www.baaa-acro.com/crash-archives?created=2015-01-02&created_1=2025-06-28&field_crash_region_target_id=All&field_crash_country_target_id=&field_crash_registration_target_id=&field_crash_aircraft_target_id=&field_crash_operator_target_id=&field_crash_cause_target_id=All&field_crash_zone_target_id=&field_crash_site_type_target_id=All&field_crash_phase_type_target_id=All&field_crash_flight_type_target_id=All&field_crash_survivors_value=All&field_crash_city_target_id=&page="
df = pd.read_html(url)[0]
```

```python
df.head()
```

#### There are 67 pages of results for this range of dates, so I'm going to scrape every page using a for loop.

```python
all_tables = []

for page_num in range(67):
    full_url = url + str(page_num)
    try:
        df = pd.read_html(full_url)[0]
        all_tables.append(df)
        print(f"Page {page_num + 1} scraped successfully.")
    except Exception as e:
        print(f"Failed to scrape page {page_num + 1}: {e}")
```

#### Now, I'm going to combine all 67 pages into one large table to easily analyze the data!

```python
final_df = pd.concat(all_tables, ignore_index=True)
```

```python
extra_final_df = final_df.drop(columns=['Image', 'Registration', 'Unnamed: 7'])
```

#### I want to categorize the aircraft into 5 categories, so I'm grouping every aircraft by their string names in the following categories.

```python
def categorize_aircraft(ac_type):
    if pd.isna(ac_type):
        return 'Unknown'
    ac_type = ac_type.lower()
    if any(x in ac_type for x in ['cessna', 'piper', 'beechcraft', 'pilatus', 'de havilland', 'king air']):
        return 'Small Propeller Plane'
    elif any(x in ac_type for x in ['boeing', 'airbus', 'mcdonnell', 'embraer']):
        return 'Commercial Jet'
    elif any(x in ac_type for x in ['hercules', 'antonov', 'ilyushin', 'c-130', 'an-', 'mil']):
        return 'Military / Cargo'
    elif 'helicopter' in ac_type:
        return 'Helicopter'
    else:
        return 'Other'
```

```python
extra_final_df['Aircraft Category'] = extra_final_df['A/C Type'].apply(categorize_aircraft)
```

```python
category_counts = extra_final_df['Aircraft Category'].value_counts().reset_index()
category_counts.columns = ['Category', 'Number of Accidents']
```

```python
category_counts.to_csv('aircraft_categories_for_datawrapper.csv', index=False)
```
